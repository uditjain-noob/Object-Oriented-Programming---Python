```python
import math

class Circle:
    pi = math.pi

    def __init__(self, radius):
        self.radius = radius
    
    def calcArea(self):
        return (self.radius ** 2) * self.pi
    
    def calcPeri(self):
        return 2 * self.radius * self.pi

r = int(input("Enter the radius of the circle: "))
obj1 = Circle(r)
print(f"Area of the circle of radius {obj1.radius} is: {obj1.calcArea()}")
print(f"Area of the circle of radius {obj1.radius} is: {obj1.calcPeri()}")
```

```python
class Teacher:

    def __init__(self, name, id, branch, salary):
        self.name = name
        self.id = id
        self.branch = branch
        self.salary = salary
    
    def dispInfo(self):
        print(100*"=")
        print(f"Name: {self.name}")
        print(f"ID: {self.id}")
        print(f"Branch: {self.branch}")
        print(f"Salary: {self.salary} Rupees")

    @classmethod
    def createTeacher(self):
        print(100*"=")
        name = input("Enter the name of the teacher: ")
        id = input("Enter the id of the teacher: ")
        branch = input("Enter the branch name of the teacher: ")
        salary = int(input("Enter the salary of the teacher: "))
        return Teacher(name, id, branch, salary)

n = int(input("Enter the number of teachers you would like to input: "))
listofTeachers = []
for i in range(n):
    listofTeachers.append(Teacher.createTeacher())
print("The information of all the teachers is: ")
for i in listofTeachers:
    i.dispInfo()
```

```python
class Account:
    def __init__(self, id, name, amount):
        self.id = id
        self.name = name
        self.amount = amount

    def deposit(self, depAmount):
        print(f"You deposited {depAmount} Rupees")
        self.amount += depAmount

    def withdraw(self, withAmount):
        if self.amount >= withAmount:
            self.amount -= withAmount
            print(f"You withdrew {withAmount} rupees from your account")
            return True
        print(f"You do not have enough balance in your account to withdraw {withAmount}")
        return False

    def checkBal(self):
        return self.amount

    def dispAcc(self):
        print(100*"=")
        print(f"Account Details")
        print(f"Account Number: {self.id}")
        print(f"Account Holder: {self.name}")
        print(f"Account Balance: {self.amount}")
        print(100*"=")

    @classmethod
    def createAccount(self):
        id = input("Enter the account number of the account holder: ")
        name = input("Enter the name of the account holder: ")
        amount = int(input("Enter the amount already present in the account: "))
        return Account(id, name, amount)
        
acc1 = Account.createAccount()
acc1.dispAcc()
depamount = int(input("Enter the amount you would like to deposit: "))
acc1.deposit(depamount)
acc1.dispAcc()
withamount = int(input("Enter the amount of money you would like to withdraw: "))
acc1.withdraw(withamount)
acc1.dispAcc()
```

```python
class Student:

    def __init__(self, name, rollno, marks):
        self.name = name
        self.rollno = rollno
        self.marks = marks
    
    def display(self):
        print(100*"=")
        print("Details of the student")
        print(f"Name: {self.name}")
        print(f"Roll No.: {self.rollno}")
        print("Marks in subjects: ")
        print(f"\tPhysics: {self.marks[0]}")
        print(f"\tChemistry: {self.marks[1]}")
        print(f"\tBiology: {self.marks[2]}")
    
    @classmethod
    def createStudent(self):
        print(100*"=")
        name = input("Enter the name of the student: ")
        rollno = input("Enter the rollno of the student: ")
        phymarks = int(input("Enter the marks awarded in Physics: "))
        chemmarks = int(input("Enter the marks awarded in Chemistry: "))
        biomarks = int(input("Enter the marks awarded in Biology: "))
        return Student(name, rollno, [phymarks, chemmarks, biomarks])

s1 = Student.createStudent()
s1.display()
```

```python
class Patient:
    def __init__(self, name, admission, symptoms, oxylevel, discharge):
        self.name = name
        self.oxylevel = oxylevel
        self.symptoms = symptoms
        self.admission = admission
        self.discharge = discharge
    
    @classmethod
    def createPatient(self):
        print(100*"=")
        name = input("enter the name of the patient: ")
        oxylevel = int(input("enter the oxygen level of the patient: "))
        symptoms = input("Enter the symptoms split by a space: ").split(" ")
        admission = input("Enter the date of admission: ")
        discharge = input("Enter the date of discharge: ")
        return Patient(name, admission, symptoms, oxylevel, discharge)
    
    def dispPatient(self):
        print(100*"=")
        print(f"The Patients Information: ")
        print(f"Name: {self.name}")
        print(f"Oxygen Level: {self.oxylevel}")
        print(f"Admitted at: {self.admission}\tDischarged at {self.discharge}")
        print(f"The symptoms of the patient are {self.symptoms}")
    
class Hospital:
    oxybase = 90
    def __init__(self, listofPatients = None):
        if listofPatients == None:
            self.listofPatients = []
        else:
            self.listofPatients = listofPatients
    
    def addPatient(self):
        self.listofPatients.append(Patient.createPatient())
    
    def checkOxy(self):
        count = 0
        for pat in self.listofPatients:
            if pat.oxylevel < self.oxybase:
                count += 1
        return count
    
    def generalward(self):
        return len(self.listofPatients) - self.checkOxy()
    
    def totNum(self):
        return len(self.listofPatients)
    
    def checkSeverity(self, name):
        for i in range(self.totNum()):
            if self.listofPatients[i].name == name:
                break
        else:
            print(f"{name} is not found in the list of patients")
            return
        
        if self.listofPatients[i].oxylevel < 90:
            print(f"{name} is in a severe condition")
        else:
            print(f"{name} is not in a severe condition")
    
    def dispInfo(self):
        for i in self.listofPatients:
            i.dispPatient()
        
hosp1 = Hospital()
hosp1.addPatient()
hosp1.addPatient()
hosp1.addPatient()
hosp1.dispInfo()
print(f"The hospital has {hosp1.totNum()} Patients")
print(f"The hospital has {hosp1.checkOxy()} Patients with low oxygen levels")
print(f"The hospital has {hosp1.generalward()} Patients in the general ward")
''
```

```python
class Time:

    def __init__(self, hours = 0, minutes = 0, seconds = 0):

        self.minutes = seconds // 60
        self.seconds = seconds % 60
        self.minutes += minutes % 60
        self.hours = self.minutes // 60
        self.hours += hours
        self.hours %= 24

    def disp(self):
        print(f"{self.hours}:{self.minutes}:{self.seconds}")

hrs = int(input("Enter the hours: "))
mins = int(input("Enter the minutes: "))
secs = int(input("Enter the seconds: "))
t1 = Time(hrs, mins, secs)
t1.disp()
```

```python
class Student:

    def __init__(self, name, age, rollno):
        self.name = name
        self.age = age
        self.rollno = rollno
    
    @classmethod
    def compare(cls, first, other):
        if first.age == other.age:
            print("The age is equal for both the students")
        else:
            print("The ages are not equal of the students")
    
s1 = Student("Rahul", 13, 61)
s2 = Student("Amit", 14, 1)
Student.compare(s1, s2)
```

```python
class Laptop:
    def __init__(self, cpu, ram , disk, display):
        self.cpu = cpu
        self.ram = ram
        self.disk = disk
        self.display = display
    
    def printDetails(self):
        print("The specifications for the laptop are: ")
        print(f"CPU: {self.cpu}")
        print(f"RAM: {self.ram} GB")
        print(f"Disk Space: {self.disk} GB")
        print(f"Display (FrameRate): {self.display}Hz")

    @classmethod
    def createLaptop(self):
        print(100*"=")
        print("Enter the specifications of the laptop owned by the student: ")
        cpu = input("Enter the cpu specifications: ")
        ram = int(input("Enter the amount of ram the laptop has (GB)"))
        disk = int(input("Enter the aount of disk space the laptop holds (GB)"))
        display = int(input("Enter the frame rate of the laptop(Hz): "))
        return Laptop(cpu, ram, disk, display)
    
class Student:
    def __init__(self, name, rollno, sem, laptops = None):
        self.name = name
        self.rollno = rollno
        self.sem = sem
        if laptops == None:
            self.laptops = []
        else:
            self.laptops = laptops
        
    def dispStudent(self):
        print(100*"=")
        print(f"Name: {self.name}")
        print(f"Roll No.: {self.rollno}")
        print(f"Semester: {self.sem}")
        print(f"Laptops Owned With Specifications: ")
        for i in self.laptops:
            print(100*"=")
            i.printDetails()
    def addLaptop(self):
        self.laptops.append(Laptop.createLaptop())
    @classmethod
    def createStudent(self):
        print(100*"=")
        name = input("Enter the name of the student: ")
        rollno = input("Enter the rollno of the student: ")
        sem = int(input("Enter which semester the student is in: "))
        return Student(name, rollno, sem)

s1 = Student.createStudent()
s1.addLaptop()
s1.addLaptop()
s1.addLaptop()
s1.dispStudent()



```
